```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("europaenergy_today.csv")

# And remove the column with local currency – it is always missing
df = df.drop(['price(currency/mwh)'], axis=1)
```

```python
df1 = round(df.groupby('country')['price(eur/mwh)'].mean())
```

```python
df1.columns = ['country', 'price', 'date']
df1
```

```python
df1.to_csv('today_price_long.csv', index=True, header=True)
```

```python
df = pd.read_csv('today_price_long.csv')
df['date'] = pd.to_datetime('today').strftime("%m/%d/%Y")
df
```

```python
#Reorganize them
df1 = df[['date', 'country', 'price(eur/mwh)']]
df1
```

```python
df1.to_csv("prices_long.csv", mode='a', index=True, header=False)
```

```python
df2 = df1.pivot_table(index = 'date', columns = 'country', values = 'price(eur/mwh)', aggfunc = np.mean, dropna=False)
df2
```

```python
df2.to_csv("price_wide.csv", mode ='a', index=True, header=False)
```
